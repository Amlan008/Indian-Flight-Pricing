```python
class_1_data = df[df['class'] == 'Economy']

# Group the filtered data by 'airline' and calculate the mean price
mean_prices_class_1 = class_1_data.groupby('airline')['price'].mean().reset_index()

# Define a custom red color palette
custom_palette = sns.color_palette("Reds", len(mean_prices_class_1))

# Plot the mean prices for class 1
plt.figure(figsize=(10, 6))
sns.barplot(x='airline', y='price', data=mean_prices_class_1, palette=custom_palette)
plt.title('Mean Price for Economy Class for Each Airline')
plt.xlabel('Airline')
plt.ylabel('Mean Price')
plt.xticks(rotation=45)

plt.savefig('mean_price_economy_class.png', dpi=500)  # Adjust the filename and DPI as needed
plt.show()
```
